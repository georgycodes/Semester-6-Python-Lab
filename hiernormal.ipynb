```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Create dataset
data = {
    'Likes': [100, 120, 300, 320, 500, 520],
    'Shares': [10, 15, 50, 55, 80, 85],
    'Comments': [5, 8, 20, 22, 40, 42],
    'Posts': [2, 3, 6, 7, 10, 11]
}

df = pd.DataFrame(data)

print(df)

linked = linkage(df, method='ward', metric='euclidean')

plt.figure(figsize=(12,6))
dendrogram(linked)
plt.title("Dendrogram")
plt.xlabel("Users")
plt.ylabel("Euclidean Distance")
plt.show()

agglo = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
df['Cluster'] = agglo.fit_predict(df)

print(df)
```
