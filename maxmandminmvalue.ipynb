```python
list=[11,35,67,23,45,5,78,90,100,110]
minval=list[0]
maxval=list[0]
for i in range(0,len(list)-1):
    if minval>list[i]:
        minval=list[i]
print(f"Minimum value is {minval}")

for i in range(0,len(list)):
    if maxval<list[i]:
        maxval=list[i]
print(f"Maximum value is {maxval}")
```
