```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create dataset
data = {
    'Mathematics': [85, 78, 90, 45, 50, 60, 95, 40],
    'Programming': [88, 75, 92, 50, 48, 62, 96, 42],
    'DataAnalytics': [84, 80, 91, 47, 52, 65, 94, 45]
}

df = pd.DataFrame(data)

print(df)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

print("Cluster Centers:")
print(kmeans.cluster_centers_)

# Scatter plot
plt.scatter(df['Mathematics'], df['Programming'], c=df['Cluster'])
plt.xlabel("Mathematics")
plt.ylabel("Programming")
plt.title("K-Means Clustering")
plt.show()

print(df)
```
