```python
n = int(input("Enter number of lines: "))

text = []
print("Enter the text lines:")
for i in range(n):
    line = input()
    text.append(line)

def mapper(text):
    mapped = []
    for line in text:
        words = line.split()
        for word in words:
            mapped.append((word.lower(), 1))
    return mapped

def shuffle_sort(mapped_data):
    shuffled = {}
    for word, count in mapped_data:
        if word not in shuffled:
            shuffled[word] = []
        shuffled[word].append(count)
    return shuffled

def reducer(shuffled_data):
    reduced = {}
    for word, counts in shuffled_data.items():
        reduced[word] = sum(counts)
    return reduced

mapped_data = mapper(text)
shuffled_data = shuffle_sort(mapped_data)
word_count = reducer(shuffled_data)

print("\nWord Count Result:")
for word, count in word_count.items():
    print(word, ":", count)
```
